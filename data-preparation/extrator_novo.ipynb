```python
import youtube_dl
import pandas as pd
import numpy as np
```

```python
queries = ["breakdancing", "bboy", "break dance"]
```

```python
ydl = youtube_dl.YoutubeDL({"ignoreerrors": True})
```

```python
resultados = []
for query in queries:
    r = ydl.extract_info("ytsearchdate1000:{}".format(query), download=False)
    for entry in r['entries']:
        if entry is not None:
            entry['query'] = query
    resultados += r['entries']
```

```python
resultados = [e for e in resultados if e is not None]
len(resultados)
```
